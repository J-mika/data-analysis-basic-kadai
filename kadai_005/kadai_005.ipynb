```python
#matplotlib.pyplotメソッドをインポート
#numpyをインポート
import matplotlib.pyplot as plt
import numpy as np

#6個のグラフを表示する領域を作成する
fig, axes = plt.subplots(2,3 , tight_layout=True)

#折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin関数
# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0,1].plot(x1, y1, linewidth=2)

#ヒストグラム
x2 = np.random.normal(15, 5, 2000)
axes[0,2].hist(x2)

#散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1,0].scatter(x3,y3)

#グラフを表示する
plt.show()
```
